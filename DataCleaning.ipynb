<b>Changes to perform :</b>
- Convert price to float
- Convert discount percentage to float
- Convert ratings to float
- Convert rating_count to float

```python
# Converting price to integer

data['discounted_price'] = data['discounted_price'].str.replace("₹",'')
data['discounted_price'] = data['discounted_price'].str.replace(",",'')
data['discounted_price'] = data['discounted_price'].astype('float64')
```

```python
data['actual_price'] = data['actual_price'].str.replace("₹",'')
data['actual_price'] = data['actual_price'].str.replace(",",'')
data['actual_price'] = data['actual_price'].astype('float64')
```

<b> Observations :</b>
- '|' is a value present in the ratings column which is categorical
- Category are divided by '|' 

```python
data['rating'] = data['rating'].str.replace('|', '-1').astype('float64')
```

```python
mean_value = np.mean(data['rating'])
```

```python
data.loc[data['rating'] == -1, 'rating'] = mean_value
```

```python
data.isna().sum()*100/len(data)
```

<b>Inference :</b> only rating_count has null values which is so small that it can be ignored 

```python
# Splitting the categories

splitted = data['category'].str.split('|',expand=True)
```

```python
splitted
```

```python
#Renaming the categories in the column
splitted = splitted.rename(columns={0:'category_1', 1:'category_2', 2:'category_3'})
```

```python
splitted.category_1.value_counts()
```

```python
splitted['category_1'] = splitted['category_1'].str.replace('&', ' & ')
splitted['category_1'] = splitted['category_1'].str.replace('OfficeProducts', 'Office Products')
splitted['category_1'] = splitted['category_1'].str.replace('MusicalInstruments', 'Musical Instruments')
splitted['category_1'] = splitted['category_1'].str.replace('HomeImprovement', 'Home Improvement')
```

```python
splitted.category_1.value_counts()
```

```python
splitted.category_2.value_counts()
```

```python
splitted['category_2'] = splitted['category_2'].str.replace('&', ' & ')
splitted['category_2'] = splitted['category_2'].str.replace(',', ', ')
splitted['category_2'] = splitted['category_2'].str.replace('HomeAppliances', 'Home Appliances')
splitted['category_2'] = splitted['category_2'].str.replace('AirQuality', 'Air Quality')
splitted['category_2'] = splitted['category_2'].str.replace('WearableTechnology', 'Wearable Technology')
splitted['category_2'] = splitted['category_2'].str.replace('NetworkingDevices', 'Networking Devices')
splitted['category_2'] = splitted['category_2'].str.replace('OfficePaperProducts', 'Office Paper Products')
splitted['category_2'] = splitted['category_2'].str.replace('ExternalDevices', 'External Devices')
splitted['category_2'] = splitted['category_2'].str.replace('DataStorage', 'Data Storage')
splitted['category_2'] = splitted['category_2'].str.replace('HomeStorage', 'Home Storage')
splitted['category_2'] = splitted['category_2'].str.replace('HomeAudio', 'Home Audio')
splitted['category_2'] = splitted['category_2'].str.replace('GeneralPurposeBatteries', 'General Purpose Batteries')
splitted['category_2'] = splitted['category_2'].str.replace('BatteryChargers', 'Battery Chargers')
splitted['category_2'] = splitted['category_2'].str.replace('CraftMaterials', 'Craft Materials')
splitted['category_2'] = splitted['category_2'].str.replace('OfficeElectronics', 'Office Electronics')
splitted['category_2'] = splitted['category_2'].str.replace('PowerAccessories', 'Power Accessories')
splitted['category_2'] = splitted['category_2'].str.replace('CarAccessories', 'Car Accessories')
splitted['category_2'] = splitted['category_2'].str.replace('HomeMedicalSupplies', 'Home Medical Supplies')
splitted['category_2'] = splitted['category_2'].str.replace('HomeTheater', 'Home Theater')
data['product_id'] = data['product_id'].str.strip()
```

```python
df = data[['product_id', 'product_name', 'category', 'discounted_price', 'actual_price', 'discount_percentage', 'rating', 'rating_count']].copy()
```

```python
df['category_1'] = splitted['category_1']
df['category_2'] = splitted['category_2']
```

```python
df.head()
```

```python
# dropping category now
df.drop(columns='category',inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
# Creating codes for ratings 

rating_score = []
for score in df['rating']:
    if score<2.0 : rating_score.append('Poor')
    elif score<3.0 : rating_score.append('Below Average')
    elif score<4.0 : rating_score.append('Average')
    elif score<5.0 : rating_score.append('Above Average')
    else : rating_score.append('Excellent')
```

```python
df['rating_score'] = rating_score
```

```python
df.describe()
```

```python
df.info()
```

```python
df['rating_score'] = df['rating_score'].astype('category')
```

```python
df['rating_score'] = df['rating_score'].cat.reorder_categories(['Below Average', 'Average', 'Above Average', 'Excellent'], ordered=True)
```

```python
#Creating Difference of Price Column between Actual Price and Discounted Price

df['difference_price'] = df['actual_price'] - df['discounted_price']
```

```python
reviewers = data[['user_id','user_name']]
reviewers
```

```python
#Splitting the strings in user_id column

reviewer_id_split = reviewers['user_id'].str.split(',', expand=False)

reviewer_id_split
```

```python
#Making user id display 1 id per row

reviewer_id_exp = reviewer_id_split.explode()

reviewer_id_clean = reviewer_id_exp.reset_index(drop=True)

reviewer_id_clean
```

```python
#Splitting the strings in user_name column

reviewer_name_split = reviewers['user_name'].str.split(',', expand=False)

reviewer_name_split
```

```python
#Making user name display 1 id per row

review_name_exp = reviewer_name_split.explode()

reviewer_name_clean = review_name_exp.reset_index(drop=True)

reviewer_name_clean
```

```python
#Creating 2 Data Frames to be merged

df21 = pd.DataFrame(data=reviewer_id_clean)
df22 = pd.DataFrame(data=reviewer_name_clean)
```

```python
#Merging the 2 dataframe containing user_id and user_name

df2 = pd.merge(df21, df22, left_index=True, right_index=True)
```
